```python
import os
import re
from Bio import SeqIO

# Define paths
alignment_fasta_path = 'Files/alignment_gRNA_sequences.fasta'
output_txt_path = 'gRNA_off_target_hits.txt'

# Function to simulate counting hits based on the provided sequences
def count_hits(sequence, genome):
    # In reality, you would align the sequence to the genome using a tool like Bowtie2 or BWA
    # Here we simulate by checking how many times the sequence appears in the genome
    return len(re.findall(sequence, genome))

# Simulated genome load (replace this with actual genome loading and alignment in real use)
# For simplicity, let's assume `hg38.fa` has been loaded into `hg38_genome` as a string
# This is an oversimplification; real genome comparison requires alignment tools
# with open('/mnt/data/hg38.fa', 'r') as f:  # Uncomment and adjust if using actual genome file
#     hg38_genome = f.read()

# Simulate hg38_genome as a small string for demo purposes
hg38_genome = "Files/hg38.fa" # Replace with actual genome data

# Parse the alignment gRNA sequences FASTA file
with open(alignment_fasta_path, 'r') as fasta_file:
    records = SeqIO.parse(fasta_file, 'fasta')
    
    with open(output_txt_path, 'w') as out_file:
        for record in records:
            sequence = str(record.seq)
            hits = count_hits(sequence, hg38_genome)
            out_file.write(f"Sequence: {sequence} {hits} Matches\n")

print(f"Off-target hit results saved to '{output_txt_path}'.")
```

```python
import os
from Bio import SeqIO
import pandas as pd

# Define paths
base_dir = "C:/Users/Bisha/Desktop/LTR2B_gRNA"
gRNA_fasta_path = os.path.join(base_dir, "Files", "alignment_gRNA_sequences.fasta")  # Ensure this is the correct file path
hg38_fasta_path = os.path.join(base_dir, "Files", "hg38.fa")
output_sam_path = os.path.join(base_dir, "Files", "alignment_gRNA.sam")

# Step 3: Parse the SAM file to count the hits for each gRNA
def count_hits(sam_file):
    hit_counts = {}
    with open(sam_file, 'r') as f:
        for line in f:
            if line.startswith('@'):  # Skip header lines
                continue
            fields = line.split("\t")
            gRNA_name = fields[0]  # QNAME field
            if gRNA_name in hit_counts:
                hit_counts[gRNA_name] += 1
            else:
                hit_counts[gRNA_name] = 1
    return hit_counts

# Step 4: Map gRNA names to sequences from the FASTA file
def map_gRNA_names_to_sequences(fasta_file):
    gRNA_map = {}
    for record in SeqIO.parse(fasta_file, "fasta"):
        gRNA_map[record.id] = str(record.seq)
    return gRNA_map

# Check if the SAM file exists
if not os.path.exists(output_sam_path):
    raise FileNotFoundError(f"{output_sam_path} does not exist. Ensure you have run the alignment step.")

# Count the hits
gRNA_hits = count_hits(output_sam_path)

# Map gRNA names to sequences
gRNA_sequences = map_gRNA_names_to_sequences(gRNA_fasta_path)

# Replace gRNA names with sequences in the hits data
gRNA_hits_with_sequences = {gRNA_sequences[gRNA_name]: hit_count for gRNA_name, hit_count in gRNA_hits.items() if gRNA_name in gRNA_sequences}

# Step 5: Display the results
gRNA_hits_df = pd.DataFrame(list(gRNA_hits_with_sequences.items()), columns=['gRNA Sequence', 'Hit Count'])
gRNA_hits_df = gRNA_hits_df.sort_values(by='Hit Count', ascending=False)

# Display the result
print(gRNA_hits_df)

# Save the result to a CSV file
gRNA_hits_df.to_csv(os.path.join(base_dir, "Files", 'gRNA_hits_summary_with_sequences.csv'), index=False)

print("gRNA hits summary saved to gRNA_hits_summary_with_sequences.csv")
```

```python
import pandas as pd

# Load the data from gRNA_hits_summary_with_sequences.csv
gRNA_hits_df = pd.read_csv("C:/Users/Bisha/Desktop/LTR2B_gRNA/Files/gRNA_hits_summary_with_sequences.csv")

# Load the cleaned alignment data into a dictionary for quick access
cleaned_alignment_hits = {}
with open("C:/Users/Bisha/Desktop/LTR2B_gRNA/Files/alignment_match_counts.txt", 'r') as f:
    for line in f:
        if line.strip():  # Ignore empty lines
            parts = line.split("\t")
            sequence = parts[0].replace("Sequence: ", "").strip()
            matches = int(parts[1].replace("Matches: ", "").strip())
            cleaned_alignment_hits[sequence] = matches

# Perform the subtraction for matching sequences
gRNA_hits_df['Cleaned Hits'] = gRNA_hits_df['gRNA Sequence'].map(cleaned_alignment_hits)
gRNA_hits_df['Difference'] = gRNA_hits_df.apply(lambda row: row['Hit Count'] - row['Cleaned Hits'] if pd.notnull(row['Cleaned Hits']) else row['Hit Count'], axis=1)

# Filter out sequences that were only in the cleaned alignment (since these should not be in the original gRNA hits)
gRNA_hits_df = gRNA_hits_df.dropna(subset=['Cleaned Hits'])

# Display the result
print(gRNA_hits_df)

# Save the result to a new CSV file
gRNA_hits_df.to_csv("C:/Users/Bisha/Desktop/LTR2B_gRNA/Files/gRNA_hits_difference.csv", index=False)
```
